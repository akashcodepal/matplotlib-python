```python
from matplotlib import pyplot as plt
# import matplotlib.pyplot as plt
import numpy as np
```

```python
x=(1,2,3,4,5,6,7,8,9)
plt.plot(x,np.sin(x))
```

```python
# linspace will create 1000 points from 0 to 10
x = np.linspace(0,10,1000)
```

```python
y=np.sin(x)
```

```python
plt.plot(x,y)
# OR plt.plot(x,np.sin(x))
```

### Another graph

```python
plt.scatter(x,y)
```

```python
# To visualise this graph with less points
# we use Slicing same as numpy
# x ==> collection of 1000 points from 0-10 ..same is y with sin function
```

```python
X = x[::10]
Y = y[::10]
```

```python
plt.scatter(X,Y)
```

## Color

```python
#color can be written full or just 1 word
b=x-2
plt.plot(x,np.sin(x),color='r')  
plt.plot(x,np.cos(x),color='Cyan')
plt.plot(x,b,'y')  #Direct color can also be Given
```

### Other Format of Color

```python
a=x+2
b=x-1
plt.plot(x,np.sin(x),'-g') #Solid Green
plt.plot(x,np.cos(x),'--c') # Dashed Cyan
plt.plot(x,a,'-.k') # DashDot Black
plt.plot(x,b,':r'); # dotted red
```

## Other Functions

```python
a=np.arange(10)
b=a+1
c=a+3
d=a+4
```

```python
plt.plot(a,b,color='r',label='B Graph')
# label will not be shown.. So use
plt.legend()
plt.title('Understand Graph')
plt.xlabel('values of A')
plt.ylabel('values of B')
plt.show()   # just use it..No matter what
```

```python
plt.plot(a,b,color='r',label='B Graph')
plt.plot(a,c,color='b',label='C Graph')
plt.plot(a,d,color='g',label='D Graph')
plt.legend()
plt.title('Drawing Multiple Grapha')
plt.xlabel('values of A')
plt.ylabel('Derived values')
plt.show()
```
